```python
lung_marker_genes = [l.strip() for l in open('lung_genes.txt').readlines()]
lung_marker_genes
```

```python
metadata = pd.read_excel('table_s1_meta_data.xlsx', sheet_name=1)
tdata = pd.read_excel('table_s2d_transcriptomic.xlsx', skiprows=2).rename({'id.1': 'transcript_name'}, axis=1)
pdata = pd.read_excel('table_s3a_proteomic.xlsx', skiprows=2).rename({'id.1': 'protein_name'}, axis=1)

samples = metadata['Sample.ID']
```

```python
from anndata import AnnData
from mudata import MuData
```

```python
tX = tdata.loc[tdata[tdata['data type'] == 'transcriptomic data'].index, tdata.columns.isin(samples)].reset_index(drop=True)
tobs = tdata.loc[tdata[tdata['data type'] == 'transcriptomic data'].index, ~tdata.columns.isin(samples)].set_index('id')
tvars = tdata.loc[tdata[tdata['data type'] == 'meta data'].index, (tdata.columns.isin(samples) | tdata.columns.isin(['id']))].set_index('id').T

pX = pdata.loc[pdata[pdata['data type'] == 'proteomic data'].index,pdata.columns.isin(samples)].reset_index(drop=True)
pobs = pdata.loc[pdata[pdata['data type'] == 'proteomic data'].index, ~pdata.columns.isin(samples)].set_index('geneSymbol')
pvars = pdata.loc[pdata[pdata['data type'] == 'meta data'].index, (pdata.columns.isin(samples) | pdata.columns.isin(['id']))].set_index('id').T
```

```python
tData = AnnData(X=tX, obs=tobs, var=tvars)
pData = AnnData(X=pX, obs=pobs, var=pvars)
mdata = MuData({"t": tData, "p": pData})
```
